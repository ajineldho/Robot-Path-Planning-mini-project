```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import heapq
import random
import pandas as pd

# Parameters
GRID_SIZE = 20
NUM_PARTS = 5
PART_TYPES = ['wire', 'switch', 'relay']
DROP_LOCATIONS = {'wire': (18, 3), 'switch': (18, 10), 'relay': (18, 17)}
OBSTACLES = [(10, i) for i in range(5, 15)]

# Generate random parts
parts = []
for _ in range(NUM_PARTS):
    x, y = np.random.randint(1, 8), np.random.randint(1, GRID_SIZE-1)
    part_type = random.choice(PART_TYPES)
    parts.append({'type': part_type, 'position': (x, y)})

# Define A* algorithm
def heuristic(a, b):
    return abs(a[0]-b[0]) + abs(a[1]-b[1])

def a_star(start, goal, obstacles):
    open_set = []
    heapq.heappush(open_set, (0, start))
    came_from = {}
    g_score = {start: 0}
    f_score = {start: heuristic(start, goal)}
    while open_set:
        _, current = heapq.heappop(open_set)
        if current == goal:
            path = [current]
            while current in came_from:
                current = came_from[current]
                path.append(current)
            return path[::-1]
        for dx, dy in [(-1,0), (1,0), (0,-1), (0,1)]:
            neighbor = (current[0]+dx, current[1]+dy)
            if 0 <= neighbor[0] < GRID_SIZE and 0 <= neighbor[1] < GRID_SIZE:
                if neighbor in obstacles:
                    continue
                tentative_g = g_score[current] + 1
                if tentative_g < g_score.get(neighbor, float('inf')):
                    came_from[neighbor] = current
                    g_score[neighbor] = tentative_g
                    f_score[neighbor] = tentative_g + heuristic(neighbor, goal)
                    heapq.heappush(open_set, (f_score[neighbor], neighbor))
    return []

# Robot simulation loop
robot_position = (0, 0)
paths = []
data_records = []

for part in parts:
    part_pos = part['position']
    drop_pos = DROP_LOCATIONS[part['type']]
    path_to_part = a_star(robot_position, part_pos, OBSTACLES)
    path_to_drop = a_star(part_pos, drop_pos, OBSTACLES)
    full_path = path_to_part + path_to_drop[1:]
    paths.append((part['type'], full_path))
    robot_position = drop_pos
    data_records.append([part['type'], part_pos, drop_pos])

# Visualize the simulation
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_xlim(0, GRID_SIZE)
ax.set_ylim(0, GRID_SIZE)
ax.set_xticks(range(0, GRID_SIZE))
ax.set_yticks(range(0, GRID_SIZE))
ax.grid(True)

# Draw obstacles
for ox, oy in OBSTACLES:
    ax.add_patch(patches.Rectangle((ox, oy), 1, 1, color='black'))

# Draw parts
for part in parts:
    x, y = part['position']
    ax.add_patch(patches.Circle((x+0.5, y+0.5), 0.3, color='blue'))
    ax.text(x+0.1, y+0.1, part['type'][0].upper(), fontsize=8, color='white')

# Draw drop zones
for t, (x, y) in DROP_LOCATIONS.items():
    ax.add_patch(patches.Rectangle((x, y), 1, 1, color='green'))
    ax.text(x+0.1, y+0.1, t[0].upper(), fontsize=8, color='white')

# Draw paths
for label, path in paths:
    xs, ys = zip(*[(x+0.5, y+0.5) for x, y in path])
    ax.plot(xs, ys, label=label)

ax.legend()
plt.title("Simulated Robot Path Planning for Automated Sorting")
plt.tight_layout()
plt.show()

# Show data table
df = pd.DataFrame(data_records, columns=["Part Type", "Start Position", "Drop Location"])
print(df)
```
